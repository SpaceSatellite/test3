```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('CSV_ETH/ETHUSDT-15m-2025-01.csv')
print(list(data))
```

```python
eth_open = data['open']
eth_high = data['high']
eth_low = data['low']
eth_close = data['close']
```

```python
MA = {}
for i in range(1,101):
    ma = []
    for k in range(len(eth_close)-i):
        ma.append(sum(eth_close[0+k:i+k])/i)
    MA[i] = ma
```

```python
table = MA.items()
```

```python
for i in table:
    print(i)
    print('------------------------------------------')
```
